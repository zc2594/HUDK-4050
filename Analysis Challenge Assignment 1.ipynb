### calculate Simpsion's Index

```python
race_info
```

```python
data[race_info].info()
```

Exlcude UGDS_WHITENH, UGDS_BLACKNH, UGDS_API, UGDS_AIANOld, UGDS_HISPOld since they only have NaN

```python
data.drop(columns=['UGDS_WHITENH', 'UGDS_BLACKNH', 'UGDS_API', 'UGDS_AIANOld', 'UGDS_HISPOld'], inplace=True)
```

```python
# update race info
race_info = [i  for i in data.columns if 'UGDS_' in i]
```

```python
# function to calculate Simpsion index
def simpsion_index(races):
    num_col = len(race_info)
    sum_sqr = 0
    for i in race_info:
        sum_sqr += races[i]**2
    return sum_sqr
    
```

```python
data['SI'] = data.apply(simpsion_index,axis=1)
```

```python
data.head(3)
```

```python
# update data dataframe again

data = data.loc[:,['INSTNM','CITY','STABBR','LOCALE','SAT_AVG','PFTFAC','PCIP13','total_crime','SI']]
```

```python
data.head(3)
```

### Normalize the numbers since the variables are not in the same scale

```python
# drop schools with missing values
data.dropna(axis=0,inplace=True)
```

```python
data_final = data.copy()
```

```python
data.info()
```

```python
for i in ['LOCALE','SAT_AVG','PFTFAC','PCIP13','total_crime','SI']:
    data_final[i] = (data[i] - data[i].mean())/(data[i].std())
    
```

```python
data_final
```

### The smaller the number of `LOCALE` & 'total_crime' & 'SI', the better the school, so using inverse of these variables

```python
data_final['final_score'] = -1*data_final['LOCALE'] + \
                            0.5*(data_final['SAT_AVG'] + data_final['PFTFAC']) + \
                            -1*data_final['total_crime'] + \
                            -1*data_final['SI']
### Each characteristic of school are equally important, but there are 2 variables(SAT_AVG&PFTFAC) represent the quality of school, so use the mean of them.
```

```python
data_final.sort_values(by='final_score',ascending=False).head(20)
```
